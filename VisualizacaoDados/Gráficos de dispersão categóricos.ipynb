```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
#grfico de disperso categrico
plt.figure(figsize=(10,5))
plt.title('Idade de passageiros em cada classe')
sns.stripplot(x="class", y="age", data=titanic)
plt.show()
```

```python
#grfico de enxame
plt.figure(figsize=(10,5))
plt.title('Idade de passageiros em cada classe')
sns.swarmplot(x="class", y="age", data=titanic)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Idade de passageiros em cada classe e sexo')
sns.swarmplot(x="class", y="age", hue = 'sex', data=titanic)
plt.show()
```

```python
#col, que indica as colunas da grade; e kind, que indica o tipo de visualizao. 
sns.catplot(x="pclass", y="age", hue = 'sex', col = 'survived', kind = 'swarm', data=titanic)
```
