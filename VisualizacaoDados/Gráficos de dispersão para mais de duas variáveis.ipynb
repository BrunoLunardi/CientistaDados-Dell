```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
#carrega dataset do pr√≥prio seaborn
tips = sns.load_dataset("tips")
tips.head()
```

```python
#comportamentos de cada g√™nero em rela√ß√£o √† gorjeta
plt.figure(figsize=(8,5))
plt.title('Rela√ß√£o entre valor da conta e valor da gorjeta pelo g√™nero')
sns.scatterplot(x="total_bill", y="tip", hue = 'sex', data=tips)
plt.show()
```

```python
#dia da semana √© mapeado para a cor
plt.figure(figsize=(8,5))
plt.title('Rela√ß√£o entre valor da conta e valor da gorjeta pelo dia da semana')
sns.scatterplot(x="total_bill", y="tip", hue = 'day', data=tips)
plt.show()
```

```python
#grade de visualiza√ß√£o, por dias da semana
sns.relplot(x="total_bill", y="tip", col = 'day', data=tips)
```

```python
#reta de regress√£o para duas vari√°veis
sns.lmplot(x="total_bill", y="tip", hue = 'sex', data=tips)
```

```python
#tamanho da mesa pela cor
plt.figure(figsize=(8,5))
plt.title('Rela√ß√£o entre valor da conta, valor da gorjeta e tamanho da mesa')
sns.scatterplot(x="total_bill", y="tip", hue = 'size', data=tips)
plt.show()
```

```python
#gr√°fico de bolhas
plt.figure(figsize=(8,5))
plt.title('Rela√ß√£o entre valor da conta, valor da gorjeta e tamanho da mesa')
sns.scatterplot(x="total_bill", y="tip", size = 'size', sizes=(15, 200), data=tips)
plt.show()
```

```python
#m√∫ltiplas vari√°veis do conjunto de dados
plt.figure(figsize=(12,8))
plt.title('Gorjetas')
sns.scatterplot(x="total_bill", y="tip", size = 'size', hue = 'sex', style = 'smoker', sizes=(15, 200), data=tips)
plt.show()
```
